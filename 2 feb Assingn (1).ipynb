```python
Question 1-Explain 'with' an example each when to use a 'for' loop and when to use 'while' loop?

Ans-Both 'for' loops and 'while' loops are used to execute a block of code repeatedly until a certain 
condition is met. However, there are some differences between them that make one more suitable than 
the other depending on the situation. Here are some examples:

Use a 'for' loop when:

You know in advance how many times you want to execute the loop. In this case, you can use a
for loop to iterate over a sequence of elements such as a list, tuple, or range object.
For example:
    
    # Print the numbers from 1 to 10
for i in range(1, 11):
    print(i)

    
Use a 'while' loop when:

You don't know in advance how many times you want to execute the loop, but you know the condition that
should stop the loop. In this case, you can use a 'while' loop to repeat the block of code until the condition
is no longer true.
For example:
    
    # Print the numbers from 1 to 10 using a while loop
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
Question 2- Write the python program to 'print' the sum and the product of the first 10 natural numbers by using 
'for' and 'while' loop.

Ans-Here's the Python program to print the sum and the product of the first 10 natural numbers using a for loop:

# Using for loop to calculate sum and product of first 10 natural numbers
sum = 0
product = 1

for i in range(1, 11):
    sum += i
    product *= i

print("Sum of first 10 natural numbers:", sum)
print("Product of first 10 natural numbers:", product)


And here's the Python program to print the sum and the product of the first 10 natural numbers using a while loop:


# Using while loop to calculate sum and product of first 10 natural numbers
sum = 0
product = 1
i = 1

while i <= 10:
    sum += i
    product *= i
    i += 1

print("Sum of first 10 natural numbers:", sum)
print("Product of first 10 natural numbers:", product)


Both of these programs produce the same output.
```

```python
Question 3-Create the python program to compute the electricity bill of the Household.

Ans-Here's a Python program to compute the electricity bill for a household:

# Program to compute electricity bill

# Input the number of units consumed
units = int(input("Enter the number of units consumed: "))

# Calculate the bill amount based on the units consumed

if units <= 100:
    bill = units * 4.5
elif units <= 200:
    bill = 100 * 4.5 + (units - 100) * 6
elif units <= 300:
    bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
else:
    bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20

print("Electricity Bill: Rs.", bill)


In this program, we first ask the user to 'input' the number of units consumed. Based on the number of units,
we calculate the bill amount using the following formula:

For the first 100 units: 100 * 4.5 = 450
For the next 100 units: 100 * 6 = 600
For the next 100 units: 100 * 10 = 1000
For the remaining 10 units: 10 * 20 = 200
Total bill = 450 + 600 + 1000 + 200 = 2250, which matches the expected output mentioned in the problem statement.

(corrected question)
```

```python
Question 4-Create a 'list' of numbers from 1 to 100.Use 'for' loop and 'while' loop to calculate the cube
of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list 
and 'print' that list.

Ans-Python program that creates a 'list' of numbers from 1 to 100, calculates the cube of each number
using a 'for' loop and a 'while' loop, and appends the number to a new list if the cube is divisible by 4 or 5:
    
    # Using for loop to calculate the cubes of numbers from 1 to 100
numbers = list(range(1, 101))
cubes = []

for num in numbers:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes.append(num)

print("Cubes divisible by 4 or 5 (using for loop):", cubes)


# Using while loop to calculate the cubes of numbers from 1 to 100
numbers = list(range(1, 101))
cubes = []
i = 0

while i < len(numbers):
    num = numbers[i]
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes.append(num)
    i += 1

print("Cubes divisible by 4 or 5 (using while loop):", cubes)


In this program, we first create a 'list' of numbers from 1 to 100 using the 'range()' function. 
We then use a 'for' loop and a 'while' loop to calculate the cube of each number and check if the cube 
is divisible by' 4 or 5'. If it is, we append the number to a new 'list' called cubes. Finally, we print
the 'list' of numbers whose cubes are divisible by 4 or 5.
```

```python
Question 5- Write a program to filter count vowels in the below given string.
               
                string="I want to become a Data Scientist"
        
Ans- Python program that counts the number of vowels in the given string :
    
    # Input string
string = "I want to become a Data Scientist"

# Define a function to count vowels
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count

# Call the function and print the count
vowel_count = count_vowels(string)
print("Number of vowels in the string:", vowel_count)

In this program, we first define a string that we want to count the vowels in. We then define a function 
called 'count_vowels()' that takes a string as 'input' and counts the number of vowels in the string.
We then call the 'count_vowels()' function with our input string and store the result in a variable called 
'vowel_count'. We print the 'vowel_count' variable to display the number of vowels in the input string.
```
